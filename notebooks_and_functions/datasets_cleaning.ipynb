1) Limpeza do dataset do movies100k.csv,

inicialmente ele é do tipo:

UserId | Movie | Rating | Timestamp

- Serão criados dois datsets, em um deles, os Ids serão os index e os filmes serão as colunas, as células serão preenchidas pelas avaliações do usuário para cada filme, caso o filme não tenha sido avaliado pelo usuário a célula será preenchida por Not a Number(NaN). No outro as colunas serãos usuário e as linhas serão os filmes.

- Ao final de cada dataset será criada uma coluna de médias.


```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('..\\raw_datasets\movies100k.csv', sep = '\t')
pivot_movies = movies.pivot_table(index = 'user', columns= 'movie', values= 'rating', aggfunc= 'mean')
characteristics = pd.read_csv('..\\raw_datasets\item.csv', sep = '|')
movies_names = characteristics['Movie Name'].values

for i, column in enumerate(pivot_movies.columns):
    pivot_movies.rename(columns= {column: movies_names[i]}, inplace= True)
media_rating_por_user = movies.groupby('user')['rating'].mean().reset_index()
media_rating_por_user.columns = ['user', 'user_average_rating']
pivot_movies['User Average Rating'] = media_rating_por_user['user_average_rating'].tolist()
pivot_movies.to_csv('..\cleaned_datasets\movies_pivot_1.csv', index = False)
pivot_movies
```

```python
pivot_movies = movies.pivot_table(index= 'movie', columns= 'user', values= 'rating', aggfunc= 'mean')
media_rating_por_item = movies.groupby('movie')['rating'].mean().reset_index()
media_rating_por_item.columns = ['movie', 'item_average_rating']
pivot_movies['Movie Average Rating'] = media_rating_por_item['item_average_rating'].tolist()
pivot_movies['Movie Name'] = movies_names.tolist()
pivot_movies.to_csv('..\cleaned_datasets\movies_pivot_2.csv', index= False)
pivot_movies
```

2) Limpeza do dataset de item.csv, que ja foi levemente alterado com as colunas sendo preenchidas com base no doc u.genre. A limpeza dele não envolveu a alteração da estrutura do dataset , sendo mais simples que a do dataset de movies. O processo se deu em:
- Retirar dados faltantes
- Retirar colunas que não ajudariam numa tarefa de ML
- Ajuste de uma linha que não estava na formatação correta
- Colocar apenas o mês em vez da data completa


```python
movies_ratings = pd.read_csv('..\cleaned_datasets\movies_pivot_2.csv')
characteristics = pd.read_csv('..\\raw_datasets\item.csv', sep = '|')
characteristics.drop(columns=['index','Unnamed: 3', 'imdb page'], inplace= True)
characteristics.dropna(inplace= True)
characteristics['Movie Average Rating'] = movies_ratings['Movie Average Rating']
characteristics.at[1372, "Release Date"] = '04-Feb-1971' 
characteristics['Release Date'] = characteristics['Release Date'].astype(str)
characteristics['Release Date'] = pd.to_datetime(characteristics['Release Date'], format = '%d-%b-%Y')
for index, row in characteristics.iterrows():
    characteristics.at[index,'Release Date'] = row['Release Date'].year

characteristics.rename(columns={'Release Date': 'Release Year'}, inplace= True)
characteristics.to_csv('..\cleaned_datasets\movies_genres.csv', index = False)
characteristics

characteristics
```
